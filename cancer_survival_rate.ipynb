# Data Cleaning

- Remove the `demographic_id` column. It has high cardinality and it is irrelevant.
- Remove the `year_of_death` column. It has missing values for some patients that died.
- Convert the data type of the numeric columns from `object` to `float`.
- Encode the `vital_status` and `paper_clinical_stage` columns.
- Fill missing values of `alive` patients in `days_to_death` column with a high value to signify that they are still alive.

- Apply `PCA` to reduce the dimension of the gene expression columns.

```python
# Remove Unnecessary columns
df.drop(['demographic_id', 'year_of_death'], axis=1, inplace=True)
df.head(2)
```

### Encode the `vital_status` and `paper_clinical_stage` columns

**Vital Status Encoding**

Alive ---> 0

Dead ----> 1

**Paper Clinical Stage Encoding**

Stage I ---> 1

Stage II ----> 2

Stage III ----> 3

Stage IV ----> 4

```python
df.paper_clinical_stage = df.paper_clinical_stage.map({'Stage I': 1,
                                                       'Stage II': 2,
                                                       'Stage III': 3,
                                                       'Stage IV': 4
                                                       })

df.vital_status = df.vital_status.map({'Alive': 0,
                                       'Dead': 1
                                      })
```

```python
df.iloc[:,:7].head()
```

#### Convert the datatype of the columns from object to float

```python
df.initial_weight = df.initial_weight.astype(float)
df.age_at_diagnosis = df.age_at_diagnosis.astype(float)
df.vital_status = df.vital_status.astype(int)
df.age_at_index = df.age_at_index.astype(int)

df.days_to_birth = df.days_to_birth.astype(int)
df.year_of_birth = df.year_of_birth.astype(int)

df.initial_weight.dtype
```

```python
# Pearson's Corrrelation Plot with Seaborn
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(df.iloc[:,:6].corr(method='pearson'),
            annot=True, linewidths=.5, ax=ax);
```

* There is a negative correlation of `-1` between `days_to_birth` and `age_at_diagnosis`. The `days_to_birth` column will be removed.

* There is a negative correlation of `-0.96` between `year_of_birth` and `age_at_diagnosis`. The `year_of_birth` column will be removed.

* There is a high positive correlation, `0.96` between the `days_to_birth` and `year_of_birth`.

* There is a high positive correlation, `1.00` between the `age_at_index` and `age_at_dianosis`. The `age_at_dianosis` column will be removed.



```python
# Remove highly correlated features
df.drop(['age_at_diagnosis', 'days_to_birth', 'year_of_birth'], axis=1, inplace=True)
df.head(2)
```

The next step is to convert the data type of the gene expression columns from object to float

```python
# convert the gene expression columns data type from object to float
for i in df.iloc[:,5:].columns:
  df[i] = df[i].astype(float)

df['ENSG00000288662.1'].dtype
```

Create two separate DataFrame for the patient's clinical information and the patient's gene expression data

```python
df_clinical_info = df.iloc[:, :5]
df_gene_exp_data = df.iloc[:, 5:]

df_clinical_info.head()
```

```python
# reset index so the index starts from 0
df_clinical_info.reset_index(drop=True, inplace=True)
df_clinical_info.head()
```

```python
print("Total number of gene expression columns", len(df_gene_exp_data.columns))
df_gene_exp_data.head()
```

Use `PCA` to reduce the features of the gene expression data

```python
from sklearn.decomposition import PCA

# reduce the dimension to 50
pca = PCA(n_components=50)
pca.fit(df_gene_exp_data)
transformed = pca.transform(df_gene_exp_data)
print(transformed.shape)
```

Convert the transformed gene expression data from an array to a dataframe and assign new column names to it.

```python
columns = np.arange(50)
transformed = pd.DataFrame(transformed, columns=columns)
transformed.head()
```

```python
# Merge the patient clinical record and the transformed gene expression data
df_new = pd.concat([df_clinical_info, transformed], axis="columns")
df_new.head()
```

```python
# check if there is any missing value in the days_to_death column for dead patients
df_new.query('vital_status == 1')['days_to_death'].isnull().sum()
```

```python
# check if there is any missing value in the days_to_death column for alive patients
df_new.query('vital_status == 0')['days_to_death'].isnull().sum()
```

Patients that are alive have missing values in the days_to_death column. This makes sense because they are still alive. Removing those records will make our dataset smaller.

# Predictive Analysis

# Approach 1: 

For the first approach, I'll train several models to predict the `vital_status` of the patient. I'll use some machine learning classification algorithms. I will have to remove the `days_to_death` column to prevent data leakage.

```python
# create a copy of df_new and remove the days_to_death column
df_1 = df_new.copy()
df_1.drop(['days_to_death'], axis=1, inplace=True)
df_1.head()
```

```python
# Features Extraction
X = df_1.iloc[:, np.r_[0, 2:54]].values  # select all the index except index 1
y = df_1.vital_status.values  # vital_status is the target variable
print("Matrix of features", X, sep='\n')
print("--------------------------------------------------")
print("Target Variable", y, sep='\n')
```

### Training with the Random Forest Algorithm

I use GridSearchCV hyperparameter-tuning

Since the dataset is really small, I will use 5-fold Cross-Validation for evaluation

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

random_forest_model = RandomForestClassifier(random_state=1)

# setting different parameter combinations
parameters = [{'min_samples_split': np.arange(10,70, 5),
               'criterion': ['gini', 'entropy'],
               'n_estimators': np.arange(80,150, 10)}
              ]
# n_jobs=-1 uses all the system's processors             
grid_search = GridSearchCV(estimator=random_forest_model,
                           param_grid=parameters,
                           scoring='precision',
                           cv=5,
                           n_jobs=-1)

grid_search.fit(X, y)

# get the best precision score
best_precision = grid_search.best_score_

# get the parameters that gave the best score
best_parameters = grid_search.best_params_
rf_score = best_precision * 100
print("Best Precision score {:.2f} %".format(rf_score))
print("Best Parameters:", best_parameters)
# extract the best random forest model
best_rf_model = grid_search.best_estimator_
```

The `random forest` classifier gave a `precision` score of `61.73%`

The hyperparameters for the random forest algortihm are:
* `criterion`: entropy, 
* `min_samples_split`: 10, 
* `n_estimators`: 120

### Training with the XGBoost Algorithm

```python
from xgboost import XGBClassifier

xgboost_model = XGBClassifier(random_state=1,
                              use_label_encoder=False)  # to remove warning message

# setting different parameter combinations
parameters = [{'learning_rate': np.arange(0.01, 0.5),
               'max_depth':np.arange(3, 10, 2) ,
               'n_estimators': np.arange(200, 500, 50)
               }]

# n_jobs=-1 uses all the system's processors             
grid_search = GridSearchCV(estimator=xgboost_model,
                           param_grid=parameters,
                           scoring='precision',
                           cv=5,
                           n_jobs=-1)

grid_search.fit(X, y)
# get the best precision score
best_precision = grid_search.best_score_

# get the parameters that gave the best score
best_parameters = grid_search.best_params_
xgboost_score = best_precision * 100
print("Best Precision score {:.2f} %".format(xgboost_score))
print("Best Parameters:", best_parameters)

# extract the best xgboost model
best_xgboost_model = grid_search.best_estimator_
```

The `XGBoost` classifier gave a precision score of 61.39%

The hyperparameters for the XGBoost algortihm are:

* `learning_rate`: 0.01,
* `max_depth`: 5,
* `n_estimators`: 350

### Training with an MLP Algorithm

```python
from sklearn.neural_network import MLPClassifier

mlp_model = MLPClassifier(hidden_layer_sizes=(8,6),
                          max_iter=500,
                          random_state=4)

# setting different parameter combinations
parameters = [{'activation': ['logistic', 'tanh', 'relu']
              }]

# n_jobs=-1 uses all the system's processors             
grid_search = GridSearchCV(estimator=mlp_model,
                           param_grid=parameters,
                           scoring='precision',
                           cv=5,
                           n_jobs=-1)

grid_search.fit(X, y)
# get the best precision score
best_precision = grid_search.best_score_

# get the parameters that gave the best score
best_parameters = grid_search.best_params_
mlp_score = best_precision * 100
print("Best Precision score {:.2f} %".format(mlp_score))
print("Best Parameters:", best_parameters)

# extract the best mlp model
best_mlp_model = grid_search.best_estimator_
```

The `MLP` Classifier gave a precision score of `71.91%`.

The hyperparamters are:
* `activation` --> `tanh`,
* `hidden_layer_sizes`--> (8,6),
* `max_iter` --> 500

There are two hidden layers. The first hidden layer has 8 nodes, while the second hidden layer has 6 nodes.

```python
# visualize result
fig = plt.figure(figsize=(8,8))
plt.bar(np.arange(3), [rf_score, xgboost_score, mlp_score])
plt.xticks(np.arange(3), ['Random Forest', 'XGBoost', 'Multi-Layer Perceptron'])
plt.title('Precision Scores of Different Machine Learning Algorithms on the Cancer Survival Prediction')
plt.ylabel('Precision Score')
plt.xlabel('Machine Learning Algorithms');
```

The `Multi-Layer Perceptron` algorithm has the highest precision score. It is the best model. 

## Limitations
- The dataset used for this project is really small. With more data, the results of the models will significantly improve.

## Recommendations
- More data is need to train the models on.
- Other techniques like `bagging` can be used to boost the performance of the models.

# Approach 2: 

For the second approach, I'll train several models to predict the `days_to_death` of the patient. 

I'll use some machine learning regression algorithms. I will have to remove the `vital_status` column to prevent data leakage.

The patients that are `alive` have missing data in the `days_to_death` column. This can be handled by replacing the missing data with a large value, to signify that the patients still have long to live.

Another approach will be to remove those records from the dataset. But this will significantly reduce the number of record in the dataset.

The `Root Mean Squared Error` metric will be used for evaluation.

```python
# create a copy of df_new and remove the vital_status column
df_2 = df_new.copy()
df_2.drop(['vital_status'], axis=1, inplace=True)
df_2.head()
```

Replace the missing data in the `days_to_death` column with a large value. I'm going to use `10 years*365 = 3650 days`.

```python
df_2.days_to_death.replace(np.nan, 3650, inplace=True)

# confirm if there is any missing value
print(df_2.isnull().sum().sum())

# print the shape of the dataset
print(df_2.shape)

# check the dataset
df_2.head()
```

The missing values have been replaced wih `3650`. We make the assumption that those who are alive will have 10 more years to live. 

In this approach, I don't want to remove the data of the patients that are alive. It contains useful information.

There are 57 records in our dataset

Confirm the data type of the `days_to_death` variable

```python
df_2.days_to_death.dtype
```

The `days_to_death` column is an `object`. It must be converted to an integer

```python
df_2.days_to_death = df_2.days_to_death.astype('int32')
df_2.days_to_death.dtype
```

```python
# Features Extraction
X = df_2.iloc[:, np.r_[0, 1, 3:54]].values  # select all the index except index 2
y = df_2.days_to_death.values  # days_to_death is the target variable
print("Matrix of features", X, sep='\n')
print("--------------------------------------------------")
print("Target Variable", y, sep='\n')
```

### Training with Random Forest Algorithm

For regression, the Random Forest Regressor is used.

```python
SEED = 22
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor(random_state=SEED)

# setting different parameter combinations
parameters = [{'min_samples_split': np.arange(10,70, 5),
               'n_estimators': np.arange(80,150, 10)}
              ]
# n_jobs=-1 uses all the system's processors             
grid_search = GridSearchCV(estimator=rf_regressor,
                           param_grid=parameters,
                           scoring='neg_mean_squared_error',
                           cv=5,
                           n_jobs=-1)

grid_search.fit(X, y)

# get the lowest mean squared error
# multiply by -1 to get rid of the negative number
lowest_mse = grid_search.best_score_ * -1

# get the parameters that gave the best score
best_parameters = grid_search.best_params_
rf_rmse = lowest_mse * 0.5  # get the root mean squared error
print("Lowest RMSE score {:.2f}".format(rf_rmse))
print("Best Parameters:", best_parameters)

# extract the best random forest model
rf_regressor = grid_search.best_estimator_
```

As expected, the `Root Mean Squared Error`. This is because the data is limited.

### Training with XGBoost Algorithm

```python
from xgboost import XGBRegressor

xgboost_regressor = XGBRegressor(random_state=SEED)

# setting different parameter combinations
parameters = [{'learning_rate': np.arange(0.01, 0.5),
               'max_depth':np.arange(3, 10, 2),
               'n_estimators': np.arange(200, 500, 50)
               }]

# n_jobs=-1 uses all the system's processors             
grid_search = GridSearchCV(estimator=xgboost_regressor,
                           param_grid=parameters,
                           scoring='neg_mean_squared_error',
                           cv=5,
                           n_jobs=-1)

grid_search.fit(X, y)

# get the lowest mean squared error
# multiply by -1 to get rid of the negative number
lowest_mse = grid_search.best_score_ * -1

# get the parameters that gave the best score
best_parameters = grid_search.best_params_
xgb_rmse = lowest_mse * 0.5  # get the root mean squared error
print("Lowest RMSE score {:.2f}".format(xgb_rmse))
print("Best Parameters:", best_parameters)

# extract the best XGBoost model
xgb_regressor = grid_search.best_estimator_
```

### Training with an MLP Algorithm

```python
from sklearn.neural_network import MLPRegressor

mlp_regressor = MLPRegressor(hidden_layer_sizes=(8,6),
                          max_iter=1000,
                          random_state=SEED)

# setting different parameter combinations
parameters = [{'activation': ['logistic', 'tanh', 'relu']
              }]

# n_jobs=-1 uses all the system's processors             
grid_search = GridSearchCV(estimator=mlp_regressor,
                           param_grid=parameters,
                           scoring='neg_mean_squared_error',
                           cv=5,
                           n_jobs=-1)

grid_search.fit(X, y)

# get the lowest mean squared error
# multiply by -1 to get rid of the negative number
lowest_mse = grid_search.best_score_ * -1

# get the parameters that gave the best score
best_parameters = grid_search.best_params_
mlp_rmse = lowest_mse * 0.5  # get the root mean squared error
print("Lowest RMSE score {:.2f}".format(mlp_rmse))
print("Best Parameters:", best_parameters)

# extract the best MLP model
mlp_regressor = grid_search.best_estimator_
```

```python
# visualize result
fig = plt.figure(figsize=(8,8))
plt.bar(np.arange(3), [rf_rmse, xgb_rmse, mlp_rmse])
plt.xticks(np.arange(3), ['Random Forest', 'XGBoost', 'Multi-Layer Perceptron'])
plt.title('RMSE of Different Machine Learning Algorithms on the Cancer Survival Prediction',
          fontsize=16)
plt.ylabel('RMSE')
plt.xlabel('Machine Learning Algorithms');
```

The `Random Forest` algorithm has the lowest Root Mean Squared Error. The `MLP Regressor model` has a really high Root Mean Squared Error. It couldn't converge at an iteration of 1000. Deep Learning problems require more data.
